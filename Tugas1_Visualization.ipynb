```python
# membuat dataframe dengan interval minggu
weeks = list(range(1, len(prices)+1))
df_bitcoin = pd.DataFrame({'Week': weeks, 'Price': prices})
```

```python
df_bitcoin.head(10)
```

```python
# membagikan data menjadi 2 dataframe, 2018 dan 2019
df_2018 = df_bitcoin[:52]
df_2019 = df_bitcoin[52:]
```

```python
# membuat visualisasi dengan line chart
plt.figure(figsize=(12, 6))
plt.plot(df_2018['Week'], df_2018['Price'], label="2018", marker='o')
plt.plot(df_2019['Week'], df_2019['Price'], label="2019", marker='o')
plt.title("Harga Bitcoin Mingguan (2018 vs 2019)")
plt.xlabel("Minggu")
plt.ylabel("Harga (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# melakukan perhitungan ROI dalam
price_start_2018 = df_2018['Price'].iloc[0]  # Harga awal 2018
price_end_2018 = df_2018['Price'].iloc[-1]  # Harga akhir 2018
return_2018 = ((price_end_2018 - price_start_2018) / price_start_2018) * 100
```

```python
# Menghitung return untuk tahun 2019
price_start_2019 = df_2019['Price'].iloc[0]  # Harga awal 2019
price_end_2019 = df_2019['Price'].iloc[-1]  # Harga akhir 2019
return_2019 = ((price_end_2019 - price_start_2019) / price_start_2019) * 100 
```

```python
# Menampilkan hasil
print(f"Return Tahun 2018: {return_2018:.2f}%")
print(f"Return Tahun 2019: {return_2019:.2f}%")
```

- karena data time series
- dapat melihat tren harga bitcoin serta fluktuasinya untuk melihat apakah harga bitcoin stabil atau tidak

**2. Tahun berapa pemegang bitcoin memiliki keuntungan yang lebih banyak?**

berdasarkan hasil perhitungan ROI, dan juga hasil visualisasi line chart, maka dapat disimpulkan bahwa pemegang bitcoin memiliki keuntungan yang lebih banyak pada tahun 2019. alasannya yaitu:
- harga bitcoin pada tahun 2018 berdasarkan hasil visualisasi terus mengalami penurunan yang sangat signifikan
- harga bitcoin pada tahun 2019 berdasarkan hasil visualisasi mengalami peningkatan yang signifikan terutama pada minggu ke 70-80.
- sedangkan berdasarkan hasil perhitungan ROI, harga bitcoin pada tahun 2018 memiliki return sebesar -73.51% dan tahun 2019 memiliki return sebesar 105.07% yang berarti pemegang bitcoin pada tahun 2019 memiliki keuntungan yang lebih banyak dibandingkan tahun 2018.

```python
# Nama hari dan jam
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
hours = list(range(24))

# Membuat DataFrame
df_cpu = pd.DataFrame(cpu_usage, index=days, columns=hours)
```
